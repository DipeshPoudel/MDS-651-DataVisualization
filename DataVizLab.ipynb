```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
df = pd.DataFrame({
    'Item':['Shoes','Tshirt','Pants','Phone'],
    'SLS_QTY':[100,200,300,500]
})
```

```python
df.head()
```

```python
df=df.sort_values(by=['SLS_QTY'], ascending=False)
```

```python
plt.bar(df['Item'],df['SLS_QTY'])
plt.xlabel('Items')
plt.ylabel('Sales Quantity')
```

```python
plt.pie(df['SLS_QTY'],labels=df['Item'])
plt.legend()
plt.show()
```
